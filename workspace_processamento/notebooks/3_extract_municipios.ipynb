# Imports

```python
import pandas as pd
```

# Configs

Diretório do Dicionário Oficial de Dados Enade

```python
arquivo = "..\\1_enade_original\\2023\\1.LEIA-ME\\Dicionário_arquivos_variáveis_microdados_Enade_2023.xlsx"
```

# Planilha de Municípios

```python
xls = pd.ExcelFile(arquivo)
```

Leitura do excel

```python
df = pd.read_excel(
    arquivo,
    sheet_name="MUNICÍPIOS",
    skiprows=3,
)
df.drop(columns=["Unnamed: 0"], inplace=True)
```

Renomeando e Mapeando Colunas

```python
nomes_colunas = {
    "CÓDIGO DO MUNICÍPIO": "id_municipio",
    "NOME DO MUNICÍPIO": "nome_municipio",
    "UF": "uf",
}

regioes = {
    '1': "Norte",
    '2': "Nordeste",
    '3': "Sudeste",
    '4': "Sul",
    '5': "Centro-Oeste",

}
```

```python
df.rename(columns=nomes_colunas, inplace=True)
df["regiao"] = df["id_municipio"].astype(str).str[0].map(regioes)
```

# Salvando

```python
df.to_parquet("..\\2_enade_staging\\municipios.parquet", index=False, engine='pyarrow')
```
