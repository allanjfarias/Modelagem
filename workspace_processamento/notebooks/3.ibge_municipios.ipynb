# Planilha de Municípios

```python
xls = pd.ExcelFile(arquivo)
print(xls.sheet_names)
```

```python
df = pd.read_excel(
    arquivo,
    sheet_name="MUNICÍPIOS",
    skiprows=3,
)
df.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
nomes_colunas = {
    "CÓDIGO DO MUNICÍPIO": "codigo_municipio",
    "NOME DO MUNICÍPIO": "nome_municipio",
    "UF": "uf",
}

regioes = {
    '1': "Norte",
    '2': "Nordeste",
    '3': "Sudeste",
    '4': "Sul",
    '5': "Centro-Oeste",

}
```

```python
df.rename(columns=nomes_colunas, inplace=True)
df["regiao"] = df["codigo_municipio"].astype(str).str[0].map(regioes)
print(df.columns)
print(df.head())
```

```python
df.to_csv("..\\enade_ready_to_use\\municipios.csv", index=False, encoding="utf-8")
```
