# imports

```python
import pandas as pd
import os
```

# Configurando Setup

Escolher edição e pasta de entrada 

```python
EDICAO = 2018
```

```python
PASTA_ENTRADA = f"../enade_original/{EDICAO}/DADOS/"
```

Arquivos selecionados para o projeto

```python
ARQUIVOS = [
f"microdados{EDICAO}_arq1.txt", # infos do curso 
f"microdados{EDICAO}_arq2.txt",# ensino do estudante
f"microdados{EDICAO}_arq3.txt", #NT GERAL FG CE
f"microdados{EDICAO}_arq4.txt", #QE_I63 E QE_I62
f"microdados{EDICAO}_arq5.txt", #SEXO
f"microdados{EDICAO}_arq6.txt", #IDADE
f"microdados{EDICAO}_arq8.txt", #QE_I02 -> COR/RAÇA
f"microdados{EDICAO}_arq10.txt",#QE_I04 -> ESCOLARIDADE PAI
f"microdados{EDICAO}_arq11.txt",#QE_I05 -> ESCOLARIDADE MÃE
f"microdados{EDICAO}_arq14.txt",#QE_I08 -> RENDA FAMILIAR
f"microdados{EDICAO}_arq16.txt",#QE_I10 -> HORAS DE TRABALHO POR SEMANA
f"microdados{EDICAO}_arq21.txt",#QE_I15 -> POLITICA DE COTAS
f"microdados{EDICAO}_arq23.txt", #QE_I17 -> TIPO DE ESCOLA EM QUE CURSOU O ENSINO MÉDIO
f"microdados{EDICAO}_arq27.txt", #QE_I21 -> PRIMEIRA GERAÇÃO COM ENSINO SUPERIOR
f"microdados{EDICAO}_arq31.txt", #QE_I25 -> MOTIVO ESCOLHA DO CURSO

]
```

Colunas selecionadas para o projeto

```python
COLUNAS = [
    "NU_ANO", #ANO DE REALIZAÇÃO DO ENADE
    "CO_CATEGAD",#PUBLICO OU PRIVADO
    "CO_MODALIDADE", #MODALIDADE DE ENSINO
    "CO_MUNIC_CURSO",#MUNICÍPIO DO CURSO
    "CO_UF_CURSO", #UF DO CURSO
    "CO_REGIAO_CURSO",#REGIÃO DO CURSO
    "CO_GRUPO", #NOME DE CURSO
    "ANO_FIM_EM", #ANO DE CONCLUSÃO DO ENSINO MÉDIO
    "ANO_IN_GRAD", #ANO DE INGRESSO NA GRADUAÇÃO
    "CO_TURNO_GRADUACAO",#TURNO DE GRADUAÇÃO
    "NT_GER", #NOTA GERAL
    "NT_FG", #NOTA DE FORMAÇÃO GERAL   
    "NT_CE", #NOTA DE COMPETÊNCIAS ESPECÍFICAS
    "CO_RS_I1", #DIFICULDADE PROVA FG
    "CO_RS_I2", #DIFICULDADE PROVA CE
    "TP_SEXO", #SEXO
    "NU_IDADE",#IDADE    
    "QE_I02", #COR
    "QE_I04", #ESCOLARIDADE PAI
    "QE_I05", #ESCOLARIDADE MÃE
    "QE_I08", #RENDA FAMILIAR
    "QE_I10", #HORAS DE TRABALHO POR SEMANA
    "QE_I15", #POLITICA DE COTAS
    "QE_I17", #TIPO DE ESCOLA EM QUE CURSOU O ENSINO MÉDIO
    "QE_I21", #PRIMEIRA GERAÇÃO COM ENSINO SUPERIOR
    "QE_I25", #MOTIVO ESCOLHA DO CURSO
    "QE_I62",#AVALIAÇÃO DOS EQUIPAMENTOS
    "QE_I63", #AVALIAÇÃO DOS AMBIENTES

]
```

Local de saída

```python
NOME_SAIDA = f"enade_{EDICAO}.parquet"
PASTA_SAIDA = "../enade_staging/"
os.makedirs(PASTA_SAIDA, exist_ok=True)
```

# Processamento

```python
dfs = []

for arquivo in ARQUIVOS:
    caminho = os.path.join(PASTA_ENTRADA, arquivo)
    

    try:
        df = pd.read_csv(
            caminho,
            sep=";",
            encoding="utf-8",
            usecols=lambda c: c in COLUNAS,
            low_memory=False
        )
        
        dfs.append(df)

    except Exception as e:
        print(f"Erro ao ler o arquivo {arquivo}: {e}")
```

```python
final = pd.concat(dfs, axis=1)
final = final.loc[:, ~final.columns.duplicated()]
```

# Salvar

```python
final.to_parquet(f'{PASTA_SAIDA}/{NOME_SAIDA}', index=False, engine='pyarrow')
```
