# Imports

```python
import pandas as pd
import os
pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)
```

# Configs

```python
STAGE_DIR = "..\\enade_staging"
PARQUETS = [f for f in os.listdir(STAGE_DIR) if f.endswith(".parquet")]
```

# Unindo Parquets

```python
dfs = []

for arq in PARQUETS:
    df = pd.read_parquet(os.path.join(STAGE_DIR, arq))
    dfs.append(df)
```

```python
df_enade = pd.concat(dfs, ignore_index=True)
df_enade.shape
```
