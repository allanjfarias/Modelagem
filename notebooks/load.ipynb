# imports

```python
import pandas as pd
import os
```

# Configurando Setup

```python
EDICAO = 2023
```

```python
PASTA_ENTRADA = f"./enade_raw/{EDICAO}/DADOS/"
```

```python
ARQUIVOS = [
    "microdados_enade_2017.txt",
    "microdados_enade_2018.txt",
    "microdados_enade_2019.txt"
]
```

```python
COLUNAS = [
    "NU_ANO",
    "CO_IES",
    "CO_CURSO",
    "CO_UF_CURSO",
    "NT_GER",
    "NT_FG",
    "NT_CE",
    "TP_PRES",
    "QE_I01"
]
```

```python
FORMATO_SAIDA = "parquet"   # csv ou parquet
```

```python
NOME_SAIDA = f"enade_{EDICAO}.{FORMATO_SAIDA}"
PASTA_SAIDA = "../enade_staging/"
os.makedirs(PASTA_SAIDA, exist_ok=True)
```

# Processamento

```python
dfs = []

for arquivo in ARQUIVOS:
    caminho = os.path.join(PASTA_ENTRADA, arquivo)

    try:
       
        df = pd.read_csv(
            caminho,
            sep=";",
            encoding="utf-8",
            usecols=lambda c: c in COLUNAS,
            low_memory=False
        )
    
    except Exception as e:
        print(f"Erro ao ler o arquivo {arquivo}: {e}")
        
    
    dfs.append(df)
```

```python
final = pd.concat(dfs, ignore_index=True)
```

# Salvar

```python
if FORMATO_SAIDA == "csv":
    final.to_csv(f'{PASTA_SAIDA}/{NOME_SAIDA}', index=False, encoding="utf-8")
else:
    final.to_parquet(f'{PASTA_SAIDA}/{NOME_SAIDA}', index=False)
```
