# imports

```python
import pandas as pd
import os
```

# Configurando Setup

```python
EDICAO = 2023
```

```python
PASTA_ENTRADA = f"../enade_raw/{EDICAO}/DADOS/"
```

```python
ARQUIVOS = [
"microdados2023_arq1.txt", # infos do curso
"microdados2023_arq2.txt",# ensino do estudante
"microdados2023_arq3.txt", #NT GERAL FG CE
"microdados2023_arq4.txt", #QE_I63 E QE_I62
"microdados2023_arq5.txt", #SEXO
"microdados2023_arq6.txt", #IDADE
"microdados2023_arq8.txt", #QE_I02 -> COR/RAÇA
"microdados2023_arq10.txt",#QE_I04 -> ESCOLARIDADE PAI
"microdados2023_arq11.txt",#QE_I05 -> ESCOLARIDADE MÃE
"microdados2023_arq14.txt",#QE_I08 -> RENDA FAMILIAR
"microdados2023_arq16.txt",#QE_I10 -> HORAS DE TRABALHO POR SEMANA
"microdados2023_arq21.txt",#QE_I15 -> POLITICA DE COTAS
"microdados2023_arq23.txt", #QE_I17 -> TIPO DE ESCOLA EM QUE CURSOU O ENSINO MÉDIO
"microdados2023_arq27.txt", #QE_I21 -> PRIMEIRA GERAÇÃO COM ENSINO SUPERIOR
"microdados2023_arq31.txt", #QE_I25 -> MOTIVO ESCOLHA DO CURSO

]
```

```python
COLUNAS = [
    "NU_ANO", #ANO DE REALIZAÇÃO DO ENADE
    "CO_CATEGAD",#PUBLICO OU PRIVADO
    "CO_MODALIDADE", #MODALIDADE DE ENSINO
    "CO_MUNIC_CURSO",#MUNICÍPIO DO CURSO
    "CO_UF_CURSO", #UF DO CURSO
    "CO_REGIAO_CURSO",#REGIÃO DO CURSO
    "CO_GRUPO", #NOME DE CURSO
    "ANO_FIM_EM", #ANO DE CONCLUSÃO DO ENSINO MÉDIO
    "ANO_IN_GRAD", #ANO DE INGRESSO NA GRADUAÇÃO
    "CO_TURNO_GRADUACAO",#TURNO DE GRADUAÇÃO
    "NT_GER", #NOTA GERAL
    "NT_FG", #NOTA DE FORMAÇÃO GERAL   
    "NT_CE", #NOTA DE COMPETÊNCIAS ESPECÍFICAS
    "CO_RS_L1", #DIFICULDADE PROVA FG
    "CO_RS_L2", #DIFICULDADE PROVA CE
    "TP_SEXO", #SEXO
    "NU_IDADE",#IDADE    
    "QE_I02", #COR
    "QE_I04", #ESCOLARIDADE PAI
    "QE_I05", #ESCOLARIDADE MÃE
    "QE_I08", #RENDA FAMILIAR
    "QE_I10", #HORAS DE TRABALHO POR SEMANA
    "QE_I15", #POLITICA DE COTAS
    "QE_I17", #TIPO DE ESCOLA EM QUE CURSOU O ENSINO MÉDIO
    "QE_I21", #PRIMEIRA GERAÇÃO COM ENSINO SUPERIOR
    "QE_I25", #MOTIVO ESCOLHA DO CURSO
    "QE_I62",#AVALIAÇÃO DOS EQUIPAMENTOS
    "QE_I63", #AVALIAÇÃO DOS AMBIENTES

]
```

```python
FORMATO_SAIDA = "csv"   # csv ou parquet
```

```python
NOME_SAIDA = f"enade_{EDICAO}.{FORMATO_SAIDA}"
PASTA_SAIDA = "../enade_staging/"
os.makedirs(PASTA_SAIDA, exist_ok=True)
```

# Processamento

```python
dfs = []

for arquivo in ARQUIVOS:
    caminho = os.path.join(PASTA_ENTRADA, arquivo)
    

    try:
        df = pd.read_csv(
            caminho,
            sep=";",
            encoding="utf-8",
            usecols=lambda c: c in COLUNAS,
            low_memory=False
        )

        df = df.reindex(columns=COLUNAS)
        dfs.append(df)

    except Exception as e:
        print(f"Erro ao ler o arquivo {arquivo}: {e}")
```

```python
final = pd.concat(dfs, ignore_index=True)
```

# Salvar

```python
if FORMATO_SAIDA == "csv":
    final.to_csv(f'{PASTA_SAIDA}/{NOME_SAIDA}', index=False, encoding="utf-8")
else:
    final.to_parquet(f'{PASTA_SAIDA}/{NOME_SAIDA}', index=False)
```
