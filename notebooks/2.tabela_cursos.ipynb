# Import

```python
import pandas as pd
```

# Configs

```python
ARQUIVO_TXT = "..\\enade_ready_to_use\\cursos.txt"

ARQUIVO_CSV = "..\\enade_ready_to_use\\cursos.csv"
```

# Criando Tabela

```python

dados = []

with open(ARQUIVO_TXT, "r", encoding="utf-8") as f:
    for linha in f:
        linha = linha.strip()

    
        if not linha:
            continue

    
        if "=" in linha:
            codigo, nome = linha.split("=", 1)

            codigo = codigo.strip()
            nome = nome.strip()

            # Padroniza Nome: Inicial Maiúscula em Cada Palavra
            nome = nome.title()

            dados.append({
                "codigo_curso": int(codigo),
                "nome_curso": nome
            })



```

```python

df = pd.DataFrame(dados)
df = df.drop_duplicates(subset="codigo_curso")
df = df.sort_values("codigo_curso")


print(df.head())
df.info()
df.describe()
```

```python
df.to_csv(ARQUIVO_CSV, index=False, encoding="utf-8")
```
